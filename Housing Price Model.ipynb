This is a housing price model to describe the relation among government, consumers, and real estate developers.

This rule is the same as that of hog and corn advanced by Waugh. Therefore, we can follow this idea to set up the housing supply function, as shown in Equation (1). As we all know, when housing price is too high (low) the government will take some measures to prevent price hikes (drop), thus, increasing (decreasing) the land supply is usually an effective way. Therefore, the land supply and housing price are in direct proportion, as described by Equation (2).

This means
that the real estate companies adjust their housing supply according to the relative policies
and the conditions of housing price and land price. In turn, the government regulates the land
supply according to not only the land price but also housing price.

$S^h_t = b_1p^h_t-a_1-e_1p^h_t$              (1)

$S^l_t = b_0p^l_t-a_0+e_0p^h_t$              (2)



$S^l_t$: land supply at time $t$ with positive slope in price

$S^h_t$: house supply at time $t$ with positive slope in price

$p^l_t$: land price at time $t$

$p^h_t$ housing price at time $t$





$a_0, b_0, e_0, a_1, b_1, e_1$ are positive constants.


$\frac{a_i}{b_i}(i = 0, 1)$ is the lower limit price. Suppliers provide their products only when the price reaches $\frac{a_i}{b_i}$.

Usually, the price $p$ is characterized by the nonlinear inverse demand function: $p=a-b\sqrt{Q}$, where $a, b$ are positive constants; $a$ is the maximum price in the market, and $Q$ is the total quantity in the market.

The transformation of this formula and the basic demand forms of land and housing are given by the following equations:

$D^l_t = c_0 - d_0p^l_t$                   (3)

$D^h_t = c_1 - d_1p^h_t + d_2(p^h_t)^2$         (4)

$D^l_t$: land demand at time $t$ with negative slope in price.

$D^h_t$: housing demand at time $t$ with negative slope in price.

$c_0, d_0 ,c_1, d_1, d_2$ are positive constants

$c_0, c_1$ are the quantities of saturated demand in market when price is zero.

$D^l_t = c_0 - d_0p^l_t\implies \frac{D^l_t}{d_0} = \frac{c_0}{d_0} - p^l_t$: $\frac{c_0}{d_0}$ is a upper limit price of land.

$D^h_t = c_1 - d_1p^h_t + d_2(p^h_t)^2$: Use the formula $-\frac{b}{2a}$ shown in the figure to calculate another limit of price,due to the law of demand that the slope of demand curve is negative, the demand curve shown in following figure can not go above the midpoint of the quaratic funciton,the price $p^h_t$ must satisfy the inequalities $2d_2p^h_t-d_1<0$.


These 2 limit price indicate that nobody will purchase the land until the price drops to $\frac{c_0}{d_0}$ or $\frac{d_1}{2d_2}$.

Also since the quantity cant be negative value we can get $4d_2c_1 - d^2_1>0$ by applying the discriminant equation to demand of housing. Thus, the signs of demands $D^h_t = c_1 - d_1p^h_t + d_2(p^h_t)^2$ are positive.

According to nonlinear cobweb theory, the price is determined by prior period price and
the relation of supply and demand.

$p_t = p_{t-1} + \beta (z(p_{t-1}))$ (5)

$z(p_t)=D_t-S_t$: the excess demand function is strictly decreasing and consecutive, which denotes the difference between the demand and supply.

$\beta>0$: price regulating prameter 

Substituting Equations (1)(2)(3)(4) into Equation (5):

$p^l_t=p^l_{t-1}+\beta_0(c_0-d_0p^l_{t-1}-a_0-b_0p^l_{t-1}-e_0p^h_{t-1})$

$p^h_t=p^h_{t-1}+\beta_1(c_1-d_1p^h_{t-1}+d_2(p^h_{t-1})^2-a_1-b_1p^h_{t-1}-e_1p^l_{t-1})$

$\beta_0$ measure the degree of the adjustment on the land price.

$\beta_1$ measure the degree of the adjustment on the housing price.

Due to incomplete information, consumers often stand in a wait-and-see attitude until market prices are in line with their psychological expectations. According to adaptive expectation theory, the expected price for consumers is always equal to their expectations. The expected price can be described by the following equation:

$p^e_t=p^e_{t-1}+\beta_2(p^h_{t-1}-p^e_{t-1})$


$\beta_2$ is a parameter, chosen between 0 and 1, implying that the current expectation of future price reflects the past expectation with an "error-adjustment" term, in which the current expectation increases (or decreases) according to the difference between the actual price and the previous expectation. This error-adjustment is also called 'partial adjustment'.

From Equations (6)–(8) we can get a dynamic game model about the government, real estate developers, and consumers as follows:



$p^l_t=p^l_{t-1}+\beta_0(a_2-b_2p^l_{t-1}-e_0p^h_{t-1})$


$p^h_t=p^h_{t-1}+\beta_1(c_2-e_2p^h_{t-1}+d_2(p^h_{t-1})^2+e_1p^l_{t-1})$


$p^e_t=p^e_{t-1}+\beta_2(p^h_{t-1}-p^e_{t-1})$

$a_2=c_0+a_0, b_2=d_0+b_0, c_2=c_1+a_1, e_2=d_1+b_1$

## Jacobian Matrix at $E_2$

```python
import numpy as np

J1 = Matrix([[1-2*bt0, -0.3*bt0, 0],
            [bt1*0.4, 1+2.2*bt1, 0],
            [0, bt2, 1-bt2]])

eigenvalues = J1.eigenvals()
eigenvalues
```

$\lambda_2 = 1-\beta_0+\frac{11}{10}\beta_1-\frac{\sqrt{100\beta_0^2+208\beta_0\beta_1+121\beta_1^2}}{10}$

$\lambda_1 = 1-\beta_0+\frac{11}{10}\beta_1+\frac{\sqrt{100\beta_0^2+208\beta_0\beta_1+121\beta_1^2}}{10}$

All 3 eigenvalues are real numbers since $\beta>0$ 

$\lambda_d=\lambda_1$ since $|\lambda_1|>|\lambda_2|$ and $|\lambda_1|>1>|\lambda_3|$
